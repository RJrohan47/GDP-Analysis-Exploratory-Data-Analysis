```python
df_Ind=df[df["Country Name"]=="India"]
df_Ind.head(4)
```

```python
df_Ind.plot(kind="line",x="Year",y="Value",grid=True,legend=False,figsize=(20,6),xlabel="Year",ylabel="Value_growth")
```

### Analyzing GDP growth from the data set 

```python
# to calculate gdp for each year we need to have data of current and previous year as well 
# then gdp can be calculated as 
# gdp=((curr-prev)/prev)*100
```

```python
data=df_Ind.values
data
```

```python
gdp_value=[0]

for i in range(1,len(data)):
    prev=data[i-1][3];
    curr=data[i][3];
    
    gdp_value.append(round(((curr-prev)/prev)*100,2))

    

```

```python
df_Ind=df_Ind.assign(GDP=gdp_value)
```
